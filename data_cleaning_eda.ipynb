## Handling Null Values

```python
print('Columns with null values:')
null_count = data_df.isna().sum()
null_count_df = null_count[null_count > 0].to_frame().reset_index().sort_values(0, ascending=False)
null_count_df.rename(columns={'index': 'Column', 0: 'Number of Null Values'})
```

### Columns with Null Values
* ChemicalDateRemoved
    * If applicable, date the chemical was removed from product, i.e., reformulation.
* DiscontinuedDate
    * If applicable, date the product was discontinued.
* CSF - *IGNORE COLUMN*
    * Color, scent and/or flavor as entered by manufacturer, packer and/or distributor. The color/scent/flavor name is entered as free-text into the reporting database and may contain misspellings or other errors. Not all products have specific colors, scents or flavors.
* CSFId - *IGNORE COLUMN*
    * CDPH internal identification number for a color/scent/flavor.
* BrandName 
    * Brand name as entered by manufacturer, packer and/or distributor. The brand name is entered as free-text into the reporting database and may contain misspellings or other errors.

```python
# Check data where BrandName is null
nan_brand_df = data_df[data_df['BrandName'].isna()==True]
nan_brand_df.sample(5)
```

```python
# Companies with null BrandNames
nan_brand_companies = list(nan_brand_df['CompanyName'].unique())
print(nan_brand_companies)
```

```python
# Find Companies with NaN BrandNames 
nonnull_brand_df = data_df[data_df['BrandName'].isna()==False]
some_nonnull_brand_df = nonnull_brand_df[nonnull_brand_df['CompanyName'].isin(nan_brand_companies)]
some_nonnull_brand_df['CompanyName'].unique()
```
